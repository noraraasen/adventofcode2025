```python
import pandas as pd
import math
```

```python
df = pd.read_csv('dag5.txt', delimiter=',', header=None, skiprows=0)
print(df)
```

```python
possible_cutoffs = list(range(0,len(df)))
while len(possible_cutoffs)>1:
    d = math.floor(len(possible_cutoffs)/2)
    if "-" in df.iloc[possible_cutoffs[d],0]:
        possible_cutoffs = possible_cutoffs[d:]
    else:
        possible_cutoffs = possible_cutoffs[:d]
if "-" in df.iloc[possible_cutoffs[0],0]:
    cutoff = possible_cutoffs[0]+1
else: 
    cutoff = possible_cutoffs[0]
```

```python
ranges = df.iloc[:cutoff,0]
IDs = df.iloc[cutoff:,0]
```

```python
valid_ranges = []
for i in range(len(ranges)):
    min1,max1 = ranges[i].split('-')
    valid_range = [int(min1),int(max1)+1]
    valid_ranges.append(valid_range)
```

```python
sum1 = 0
for j in range(len(IDs)):
    for i in range(len(valid_ranges)):
        checkRange = valid_ranges[i]
        checkID = int(IDs.iloc[j])
        if checkID >= checkRange[0] and checkID <= checkRange[1]:
            sum1 += 1
            break
```

```python
sum1
```

```python
valid_ranges_df = pd.DataFrame(valid_ranges, columns=['min_vals', 'max_vals'])
valid_ranges_df = valid_ranges_df.sort_values(by='min_vals').reset_index(drop=True)
```

```python
validIDlength = 0
max2 = 0
for i in range(len(valid_ranges_df)):
    min1 = valid_ranges_df.iloc[i,0]
    max1 = max(valid_ranges_df.iloc[i,1], max2)
    if i == len(valid_ranges_df)-1:
        validIDlength += max1-min1
    else:
        min2 = valid_ranges_df.iloc[i+1,0]
        if max1<=min2:
            validIDlength += max1-min1
        else:
            validIDlength += min2-min1
    max2 = max1

```
