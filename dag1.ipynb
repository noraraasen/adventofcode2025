```python
import pandas as pd
```

```python
df = pd.read_csv('dag1.txt', delimiter=';', header=None, skiprows=0)
print(df[0:10])
```

```python
## split column into one containing the first character and one containing the rest
direction = df.iloc[:,0].str[0]
value = df.iloc[:,0].str[1:]
direction = direction.replace({'L': -1, 'R': 1})
value = direction*value.astype(int)

print(value[0:10])
```

```python
ix = 50
password = 0
for i in range(len(value)):
    ix = (ix + value[i]) % 100
    if ix == 0:
        password += 1
```

```python
print("The password is: ", password)
```

```python
ix = 50
password = 0
for i in range(len(value)):
    val0 = value[i]
    if val0 < 0:
        for j in range(-val0):
            ix -= 1
            if ix % 100 == 0:
                password += 1
    elif val0 > 0:
        for j in range(val0):
            ix += 1
            if ix % 100 == 0:
                password += 1
    
```

```python
print("The password is: ", password)
```
