```python
import pandas as pd
import math
```

```python
df = pd.read_csv('dag2.txt', delimiter=',', header=None, skiprows=0).transpose()
print(df)
```

```python
badID = []

for i in range(len(df)):
    ran = df.iloc[i,0]    
    min = int(ran.split('-')[0])
    max = int(ran.split('-')[1])
    allIDs = list(range(min,max+1))
    for j in range(len(allIDs)):
        id = allIDs[j]
        l = len(str(id))
        if l % 2 == 0:
            a = int(str(id)[:(l//2)])
            b = int(str(id)[(l//2):])
            if a == b:
                badID.append(id)
                continue
        
```

```python
sum(badID)
```

```python
def split_into_n_parts(word, n):
    part_length = len(word) // n
    parts = [word[i:i + part_length] for i in range(0, len(word), part_length)]
    
    return parts
```

```python
your_list = split_into_n_parts('ababab', 3)
all(x == your_list[0] for x in your_list)
```

```python
badID = []

for i in range(len(df)):
    ran = df.iloc[i,0]    
    min = int(ran.split('-')[0])
    max = int(ran.split('-')[1])
    allIDs = list(range(min,max+1))
    for j in range(len(allIDs)):
        id = allIDs[j]
        l = len(str(id))
        k = l
        while k > 1:
            if l % k == 0:
                subIDs = split_into_n_parts(str(id), k)
                if all(x == subIDs[0] for x in subIDs):
                    badID.append(id)
                    break
            k -= 1
```
