```python
import numpy as np
```

```python
data = np.loadtxt("dag4_ex.txt", dtype = "str")
M = [list(r) for r in data]
M = np.char.replace(M, '.','0')
M = np.char.replace(M, '@','1')
M = M.astype(int)
```

```python
i,j=1,4
nrows = len(M)
ncols = len(M[0])
my_rolls = np.full((nrows, ncols), 0)

for i in range(nrows):
    for j in range(ncols):
        if M[i,j] == 1:
            
            r0 = max(i-1,0)
            r1 = min(i+1, ncols)+1
            c0 = max(j-1,0)
            c1 = min(j+1, nrows)+1

            subM = M[r0:r1,c0:c1]
            nadj = np.sum(subM)-1

            if nadj < 4:
                my_rolls[i,j] = 1
```
