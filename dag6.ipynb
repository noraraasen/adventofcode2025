```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('dag6.txt', delimiter=r'\s+', header=None, skiprows=0)
print(df)
```

```python
nrows = df.shape[0]
total = 0 
for i in range(df.shape[1]):
    int_list = [int(item) for item in df.iloc[0:nrows-1,i]]
    if df.iloc[nrows-1,i] == "*":
        answer = math.prod(int_list)
    elif df.iloc[nrows-1,i] == "+":
        answer = sum(int_list)
    total += answer
```

```python
df = pd.read_csv('dag6.txt', delimiter=r'\s+', header=None, skiprows=0)
operators = df.iloc[df.shape[0]-1,]
matrix_np = np.loadtxt('dag6.txt', dtype=str, delimiter = ",") 
```

```python
M = []
for i in range(len(matrix_np)-1):
    M.append(list(matrix_np[i]))
M = np.array(M)
```

```python
nrows = M.shape[0]
ncols = len(operators)
numbers = []
new_numbers = []
divider = ' '*nrows

row_i = 0
col_j = 0
for i in range(M.shape[1]):
    str_digit = "".join(M[:,i])
    if str_digit == divider:
        numbers.append(new_numbers)
        new_numbers = []
        row_i = 0
        col_j += 1
    else:
        new_numbers.append(str_digit)
        row_i += 1
    if(i == M.shape[1]-1):
        numbers.append(new_numbers)
```

```python
print(len(operators))
print(len(numbers))
```

```python
total = 0 
for i in range(len(numbers)):
    int_list = [int(item) for item in numbers[i]]
    if operators[i] == "*":
        answer = math.prod(int_list)
    elif operators[i] == "+":
        answer = sum(int_list)
    total += answer
```
