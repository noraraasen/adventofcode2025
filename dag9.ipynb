```python
import numpy as np
```

```python
data = np.loadtxt("dag9.txt", delimiter=',')
nrows = data.shape[0]
```

```python
rect_M = np.full([nrows, nrows], np.nan)
for i in range(nrows):
    if(i == 0):
        next
    for j in range(i):
        x = data[i, ]
        y = data[j, ]
        d1 = abs(x[0]-y[0])+1
        d2 = abs(x[1]-y[1])+1
        rect_M[i,j] = d1*d2
print(rect_M[0:4,0:4])
```

```python
np.nanmax(rect_M)
```

```python
done = False
while not done:
    idx = np.nanargmax(rect_M)
    ix1, ix2 = np.unravel_index(idx, rect_M.shape)
    rect_M[ix1, ix2] = np.nan
    coord1 = data[ix1]
    coord2 = data[ix2]
    xmin = min(coord1[0],coord2[0])
    ymin = min(coord1[1], coord2[1])
    xmax = max(coord1[0],coord2[0])
    ymax = max(coord1[1], coord2[1])
    point_one = False
    point_two = False
    if all(coord1 == [xmin,ymin]) or all(coord2 == [xmin,ymin]):
        # da ligger et punkt i øvre venstre kvadrant og et i nedre høyre
        if any((data[:,0] <= xmin) & (data[:,1] >= ymax)):
            point_one = True
        if any((data[:,0] >= xmax) & (data[:,1] <= ymin)):
            point_two = True
    else:
        # da ligger et punkt i nedre venstre kvadrant og et i øvre høyre
        if any((data[:,0] <= xmin) & (data[:,1] <= ymin)):
            point_one = True
        if any((data[:,0] >= xmax) & (data[:,1] >= ymax)):
            point_two = True
    if point_one & point_two:
        done = True
```

```python
print(data[(data[:,0] <= xmin) & (data[:,1] <= ymin)])
print(data[(data[:,0] >= xmax) & (data[:,1] >= ymax)])
```
