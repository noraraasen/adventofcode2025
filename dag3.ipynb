```python
import pandas as pd
```

```python
df = pd.read_csv('dag3.txt', delimiter=';', header=None, skiprows=0)
print(df[0:10])
```

```python
def find(s, ch):
    return [i for i, ltr in enumerate(s) if ltr == ch]
```

```python
jolt_list = []
jolt = 0

for i in range(len(df)):
    bank = str(df.iloc[i,0])
    big_last_digit = False
    for j in range(10):
        digit = 10-j
        if str(digit) in bank:
            big_digit = find(bank, str(digit))
            if len(big_digit) == 1 and big_digit[0] == (len(bank)-1):
                ones_digit = digit
                big_last_digit = True
            else:
                break
    tens_digit = digit
    if big_last_digit:
        jolt_list.append(int(str(tens_digit) + str(ones_digit)))
        jolt += int(str(tens_digit) + str(ones_digit))
    elif len(big_digit)>1:
        ones_digit = tens_digit
        jolt_list.append(int(str(tens_digit) + str(ones_digit)))
        jolt += int(str(tens_digit) + str(ones_digit))
    elif len(big_digit)==1:
        for j in range(tens_digit-1):
            digit = tens_digit-j-1
            if str(digit) in bank[big_digit[0]:]:
                ones_digit = digit
                jolt_list.append(int(str(tens_digit) + str(ones_digit)))
                jolt += int(str(tens_digit) + str(ones_digit))
                break
```

```python
jolt_list = []
jolt = 0

for i in range(len(df)):
    bank = str(df.iloc[i,0])
    banklist = [int(b) for b in list(bank)]
    big_digit = ''
    for j in range(12):
        if j == 11:
            digit = max(banklist)
        else:
            digit = max(banklist[:-11+j])
        big_digit = big_digit + str(digit)
        banklist = banklist[(banklist.index(digit)+1):]
    jolt_list.append(big_digit)
    jolt += int(big_digit)
```
