```python
import numpy as np
```

```python
input = np.loadtxt('dag7.txt', dtype=str, delimiter = ",") 
M = []
for i in range(len(input)-1):
    M.append(list(input[i]))
M = np.array(M)
```

```python
new_list_of_indexes = np.where(M == 'S')[1].tolist()
nrows = M.shape[0]-1
splits = 0
for i in range(nrows):
    list_of_indexes = list(set(new_list_of_indexes))
    new_list_of_indexes = []
    for j in range(len(list_of_indexes)):
        ix = list_of_indexes[j]
        if M[i+1,ix] == '^':
            new_list_of_indexes.append(ix-1)
            new_list_of_indexes.append(ix+1)
            splits += 1
        else:
            new_list_of_indexes.append(ix)
```
